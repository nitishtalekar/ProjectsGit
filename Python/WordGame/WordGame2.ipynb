```python
import pymysql
conn = pymysql.connect(host="localhost",user="root",passwd="",db="wordgame")
cur = conn.cursor()

try:
    d = "select * from words"
    cur.execute(d)
    data = cur.fetchall()
    my_dict = {'Words':[i[1] for i in data],'Hintone':[i[2] for i in data],'Hinttwo':[i[3] for i in data],'Hintthree':[i[4] for i in data]}
    print(my_dict)
except Exception as e:
    print("ERROR=",e)
    
WordList = my_dict['Words']
Hint1=my_dict['Hintone']
Hint2=my_dict['Hinttwo']
Hint3=my_dict['Hintthree']
WordList1 = my_dict['Words']+[0,0]
```

```python
from random import *

org = "  ."
Index = -1
opt = 0
Points = 0
counter = 0

def newword():
    shuffle(WordList)
    global org 
    org = WordList[0]
    x=list(WordList[0])
    global Index
    Index = WordList1.index(org)
    shuffle(x)
    word = "".join(x)
    return word

words = newword()

def Clean():
    l5.config(text="")
    l6.config(text="")
    textbox.delete(0,END)
    h1.config(text="")
    h2.config(text="")
    h3.config(text="")

def CleanAll():
    l5.config(text="")
    l6.config(text="")
    h1.config(text="")
    h2.config(text="")
    h3.config(text="")
    pts.config(text="")
    HintBtn1.destroy()
    HintBtn2.destroy()
    HintBtn3.destroy()
    SubmitButton.destroy()
    textbox.destroy()

import tkinter 
tk=tkinter.Tk()
tk.title("GUESS ME!!")
tk.geometry('500x550+80+80')
tk.configure(background='lightskyblue')
from tkinter import *


def submit():
    global Points
    global opt
    global counter
    if counter == len(WordList1)-2:
        CleanAll()
        l1=Label(tk,text="YOU WIN",bg="Red", font=("Helvetica", 35))
        l1.place(x=155,y=120) 
        l2=Label(tk,text="FINAL POINTS = "+str(Points),fg="Blue",bg='lightskyblue',font=("Helvetica", 20))
        l2.place(x=150,y=255)
    else:
        mtext=typword.get()
        if mtext == org:
            opt = 0
            counter = counter+1
            Points = Points+5
            WordList.remove(org)
            Clean()
            if counter == len(WordList1)-2:
                submit()
            else:
                k = newword()
                wrd.config(text=k)
                pts.config(text="POINTS = "+str(Points)+" ")
        else:
            l5.config(text="WRONG GUESS : USE HINT!")
            l6.config(text="ONE POINT DEDUCTED")
            Points = Points-1
            pts.config(text="POINTS = "+str(Points)+" ")
    
typword=StringVar()

wrd=Label(tk,text=words,bg="Yellow",font=20)
wrd.place(x=175,y=120)

pts=Label(tk,text="POINTS = "+str(Points)+" ",bg='lightskyblue',font=10)
pts.place(x=360,y=50)

textbox=Entry(tk,textvariable=typword)
textbox.place(x=175,y=160)

SubmitButton=Button(tk,text="SUBMIT",fg="Red",font=16)
SubmitButton.place(x=200,y=200)
SubmitButton.config(command=submit)

l5=Label(tk,text="",fg="Blue",bg='lightskyblue',font=16)
l5.place(x=125,y=250)
l6=Label(tk,text="",fg="Blue",bg='lightskyblue',font=16)
l6.place(x=150,y=275)
h1=Label(tk,text="",fg="Red",bg='lightskyblue',font=20)
h1.place(x=20,y=420)
h2=Label(tk,text="",fg="Red",bg='lightskyblue',font=20)
h2.place(x=20,y=450)
h3=Label(tk,text="",fg="Red",bg='lightskyblue',font=20)
h3.place(x=20,y=480)
    
def hint1():
    h1.config(text=Hint1[Index])
    global Points
    Points = Points-1
    pts.config(text="POINTS = "+str(Points)+" ")
    global opt 
    opt = 1
    
def hint2():
    global opt
    if opt == 1:
        h2.config(text=Hint2[Index])
        global Points
        Points = Points-1
        pts.config(text="POINTS = "+str(Points)+" ")
        opt = opt+1
    
def hint3():
    global opt
    if opt == 2:
        h3.config(text=Hint3[Index])
        global Points
        Points = Points-1
        pts.config(text="POINTS = "+str(Points)+" ")
        
HintBtn1=Button(tk,text="Hint1",fg="Blue",font=16,command=hint1)
HintBtn1.place(x=20,y=375)

HintBtn2=Button(tk,text="Hint2",fg="Blue",font=16,command=hint2)
HintBtn2.place(x=85,y=375)

HintBtn3=Button(tk,text="Hint3",fg="Blue",font=16,command=hint3)
HintBtn3.place(x=150,y=375)

tk.mainloop()
```
