```python
import requests
```

```python
r = requests.get("https://mangafast.net/bleach-chapter-641/")
# print(r.text)
```

```python
html = r.text
```

```python
x = html.split(' ')
y = []
for i in x:
    if "jpg" in i and "src" in i and "cdn" in i:
#         print(i)
        y.append(i[5:-1])
y
```

```python
import requests
jpg_list = []
for pg in range(len(y)):
#     f = open(str(pg)+'.jpg','wb')
#     f.write(requests.get(y[pg]).content)
#     f.close()
    jpg_list.append(requests.get(y[pg]).content)
jpg_list[0]
```

```python
filenames = [str(i)+".jpg" for i in range(len(y))]
filenames
```

```python
from PIL import Image

im_list = []

for i in filenames:
    im_list.append(Image.open(i))

im1 = Image.open("0.jpg")
# im2 = Image.open("/Users/apple/Desktop/bbd1.jpg")
# im3 = Image.open("/Users/apple/Desktop/bbd2.jpg")
# im_list = [im2,im3]

pdf1_filename = "bbd1.pdf"

im1.save(pdf1_filename, "PDF" ,resolution=100.0, save_all=True, append_images=im_list)
```
