```python
import pymongo
from pymongo import MongoClient
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
client = MongoClient()
db = client.ytdb
yt = db.ytdata
```

# DATA

```python
cursor = yt.find()
dat = [i for i in cursor]
dat
```

```python
len(dat)
```

# VIDEOS IN CATEGORIES

```python
c = yt.distinct('Category')
cat = [i for i in c]
cat_count = {}
for i in cat:
    x = yt.find({'Category':i}).count()
    cat_count[i] = x
```

```python
catgraph = pd.DataFrame.from_dict(cat_count,orient='index')
print(catgraph)
catgraph.plot(kind='bar',figsize=(17,10),legend=False)
plt.show()
```

# SPECIFIC YOUTUBER CHART

```python
u_name = input("ENTER YOUTUBER NAME")
```

```python
u_count = {}
for i in cat:
    x = yt.find({'Uploader':u_name,'Category':i}).count()
    if x==0:
        continue
    u_count[i] = x

u_graph = pd.DataFrame.from_dict(u_count,orient='index')
u_graph.plot.pie(y=0,figsize=(17,17))
```

# DATA ON UPLOADERS

```python
u = yt.distinct('Uploader')
up = [i for i in u]
import random as rand
rand_up = rand.sample(range(0,2900),25)
upl = []
for i in rand_up:
    upl.append(up[i])
```

```python
up_rat = {}

for i in upl:
    rat = pd.DataFrame(list(yt.find({'Uploader':i},{'Rating'})))
    rat.dropna(axis=0)
    up_rat[i] = [rat['Rating'].mean()]

urat_graph = pd.DataFrame.from_dict(up_rat,orient='index')
urat_graph.plot(kind='barh',figsize=(15,10),legend=False,title='UPLOADER RATINGS')
plt.show()    
```

```python
up_view = {}

for i in upl:
    view = pd.DataFrame(list(yt.find({'Uploader':i},{'Views'})))
    view.dropna(axis=0)
    up_view[i] = [view['Views'].sum()]

uview_graph = pd.DataFrame.from_dict(up_view,orient='index')
uview_graph.plot(kind='kde',figsize=(15,10),legend=False,title='UPLOADER VIEWINGS')
plt.show() 
```
