# PLACES

```python
wrd = []

with open('placesc.txt','r') as f:
    for line in f:
#         print(line)
        for word in line.split():
#             print(word)
            wrd.append(word)
# print(wrd)
```

```python
temp = []

for i in range(0,len(wrd)):
    if wrd[i] == "alt=":
        i = i + 2;
        l = wrd[i]
        s = " "
        i = i + 1
        while wrd[i] != '"':
#             print(l,wrd[i])
            l = l + s +wrd[i]
            i = i + 1
        temp.append(l)
```

```python
temp
```

```python
places = []
for i in range(len(temp)):
    if temp[i] == 'List Challenges':
        i = i + 1
        while i + 6 < len(temp) and temp[i + 6] != 'List Challenges':
            places.append(temp[i])
            i = i + 1

places = places[:-3]
places
```

# Animals

```python
animal = []

with open('animal.txt','r') as f:
    for line in f:
#         print(line)
        for word in line.split():
#             print(word)
            animal.append(word)
    
print(len(animal),animal)
```

```python
obj = []

with open('objevt.txt','r') as f:
    for line in f:
#         print(line)
        obj.append(''.join(list(line)[:-1]))

o = [i for i in obj if i not in animal]
objects = [i for i in o if len(i.split(" "))<=3]
print(len(obj),obj,"\n",len(o),o,objects,len(objects))
```
