```python
import pymysql
conn = pymysql.connect(host="localhost",user="root",passwd="",db="rgit_feedback")
cur = conn.cursor()

try:
    d = "select * from teacher"
    cur.execute(d)
    data = cur.fetchall()
    print(data)
except Exception as e:
    print("ERROR=",e)
```

```python
t_data = list(data)
tdata = []
for i in t_data:
    tdata.append(list(i))
tdata
```

```python
for i in tdata:
    x = "".join(i[1].split(" "))
    i[1] = x
    
tdata
```

```python
import re
for i in tdata:
    x = " ".join(re.findall('[A-Z][^A-Z]*', i[1]))
    i[1] = x

tdata
```

```python
tdict = {}
for i in tdata:
    x = i[1].split(" ")[-1]
    z = []
    for j in tdata:
        y = j[1].split(" ")[-1]
        if x == y:
            z.append(j)
    tdict[x] = z
        
```

```python
tdict
```

```python
dupl = []
non = []
for i in list(tdict.values()):
    if len(i)>1:
        dupl.append(i)
    else:
        non.append(i)
        
dupl
```

```python
d2 = dupl[:]
for i in d2:
    x = int(input(i))
    for j in i:
        if j[0] == x:
            i.remove(j)
```
