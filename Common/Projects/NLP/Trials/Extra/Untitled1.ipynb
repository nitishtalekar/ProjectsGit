```python
from googletrans import Translator

jap = []
k = 0
for i in eng:
    if i == "That's his.":
        break
    k = k + 1


for i in range(k+1,len(eng)):
    print(eng[i])
    translator = Translator()
    x = translator.translate(eng[i], dest='ja')
    print(x.text)
    jap.append(x.text)
```
