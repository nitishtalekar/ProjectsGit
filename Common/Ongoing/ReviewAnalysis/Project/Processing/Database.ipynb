```python
import pymysql
conn = pymysql.connect(host="localhost",user="root",passwd="",db="fake_feedback")
cur = conn.cursor()
```

```python
f = open("nwords.txt", "r")
x = f.read()
k = x.split(" ")

w = []
for i in k:
    if 'class=wordlist-item>' in i:
        w.append(i)
        
n = []
for i in w:
    word = i.split('>')[1][:-5]
    n.append(word)
```

```python
try:
    for i in n:
        d = "INSERT INTO negative_words (word) VALUES (%s)"
        cur.execute(d,i)
    conn.commit()
except Exception as e:
    print("ERROR=",e)
```

```python
import random as r
def ip_generator():
    num = [i for i in range(255)]
    ip = str(r.choice(num))+"."+str(r.choice(num))+"."+str(r.choice(num))+"."+str(r.choice(num))
    return ip
```

```python
import csv

input_file = csv.DictReader(open("ReviewData.csv",encoding='cp850'))
data = []

for i in input_file:
    data.append(i['Review'])
    
# data
    
ip_list=[]
for i in range(200):
    ip_list.append(ip_generator())
    
# ip_list


try:
    for i in data:
        x = [r.choice(ip_list),i]
        d = "INSERT INTO reviews(ip_address, review) VALUES (%s,%s)"
        cur.execute(d,x)
    conn.commit()
except Exception as e:
    print("ERROR=",e)

# print(data[0])
# x = [r.choice(ip_list),data[0]]
# d = "INSERT INTO reviews(ip_address, review) VALUES (%s,%s)"
# cur.execute(d,x)
```
