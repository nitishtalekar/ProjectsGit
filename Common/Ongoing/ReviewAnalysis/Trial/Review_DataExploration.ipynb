```python
f = open("amazon_reviews.txt", "r",encoding="utf8")
d = f.readlines()
print(d[:5])
```

```python
head = d[0].split('\t')
d.pop(0)
head[-1] = head[-1][:-1]
head.append('SENTIMENT')
```

```python
data = []
for i in d:
    i = i.split('\t')
    i[-1] = i[-1][:-1]
    if i[1] == '__label1__':
        i[1] = 'Fake'
    elif i[1] == '__label2__':
        i[1] = 'Not_Fake'
    if int(i[2])>3:
        i.append('Positive')
    elif int(i[2])<=3:
        i.append('Negative')
    data.append(i)
print(data[:5])
print(data[15000:15005])
```

```python
import pandas as pd
df = pd.DataFrame(data,columns=head)
df
```

```python
df.to_csv('reviews_data.csv',index=False)
```

```python
neg_fake = []
pr_fake = []
neg_notfake = []
pr_notfake = []
for index, row in df_review.iterrows():
#     print(row['REVIEW_TEXT'])
    paragraph = row['REVIEW_TEXT']
#     print(paragraph)
    count = Counter(paragraph.split())
    neg = 0
    pr = 0
    for key, val in count.items():
        key = key.rstrip('.,?!\n') # removing possible punctuation signs
        if key in negative:
            neg += val
        if key in wrd:
            pr += val
    if row['LABEL'] == 'Fake':
        neg_fake.append(neg)
        pr_fake.append(pr)
    elif row['LABEL'] == 'Not_Fake':
        neg_notfake.append(neg)
        pr_notfake.append(pr)
#     break
```
