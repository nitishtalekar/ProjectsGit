```python
f = open("amazon_reviews.txt", "r",encoding="utf8")
d = f.readlines()
print(d[:5])
```

```python
head = d[0].split('\t')
d.pop(0)
head[-1] = head[-1][:-1]
head.append('SENTIMENT')
```

```python
data = []
for i in d:
    i = i.split('\t')
    i[-1] = i[-1][:-1]
    if i[1] == '__label1__':
        i[1] = 'Fake'
    elif i[1] == '__label2__':
        i[1] = 'Not_Fake'
    if int(i[2])>3:
        i.append('Positive')
    elif int(i[2])<=3:
        i.append('Negative')
    data.append(i)
print(data[:5])
print(data[15000:15005])
```

```python
import pandas as pd
df = pd.DataFrame(data,columns=head)
df
```

```python
df.to_csv('reviews_data.csv',index=False)
```
