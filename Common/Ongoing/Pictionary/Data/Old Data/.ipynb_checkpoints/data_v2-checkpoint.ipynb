# Places

```python
wrd = []

with open('places_v2.html','r') as f:
    for line in f:
#         print(line)
        for word in line.split():
#             print(word)
            wrd.append(word)
print(wrd)
```

```python
temp = []

for i in range(0,len(wrd)):
    if wrd[i] == "alt=":
        i = i + 2;
        l = wrd[i]
        s = " "
        i = i + 1
        while wrd[i] != '"':
#             print(l,wrd[i])
            l = l + s +wrd[i]
            i = i + 1
        temp.append(l)
```

```python
temp
```

```python
places = []
for i in range(len(temp)):
    if temp[i] == 'List Challenges':
        i = i + 1
        while i + 6 < len(temp) and temp[i + 6] != 'List Challenges':
            places.append(temp[i])
            i = i + 1

places = places[:-3]
places
```

```python
len(places)
```

# Animals

```python
animal = []

with open('animal_v2.txt','r') as f:
    for line in f:
#         print(line)
#         for word in line.split():
#             print(word)
        animal.append(line[:-1])
    
print(len(animal),animal)
```

```python
food = []

with open('food_v2.txt','r') as f:
    for line in f:
#         print(line)
#         for word in line.split():
#             print(word)
        food.append(line[:-1])
food1 = list(set(food))   
# print(len(food),food)
print(len(food1),food1)
```

# Objects

```python
obj = []

with open('objevt.txt','r') as f:
    for line in f:
#         print(line)
        obj.append(''.join(list(line)[:-1]))

o = [i for i in obj if i not in animal]
objects = [i for i in o if len(i.split(" "))<=3]
print(len(obj),obj,"\n",len(o),o,objects,len(objects))
```

# Actions

```python
actions = []

with open('action.txt','r') as f:
    for line in f:
#         print(line)
        actions.append(''.join(list(line)[:-1]))

print(actions,len(actions))
```
