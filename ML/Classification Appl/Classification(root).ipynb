```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import xgboost as xgb
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np

clfa = LogisticRegression(random_state = 42)
clfb = SVC(random_state = 912, kernel = 'rbf')
clfc = xgb.XGBClassifier(seed = 2)
clfd = DecisionTreeClassifier()

clf = LogisticRegression(random_state = 42)

data = pd.DataFrame()
xtrain = pd.DataFrame()
ytrain = pd.DataFrame()
xtest = pd.DataFrame()
ytest = pd.DataFrame()

scaler = MinMaxScaler()

accu = 0
predval = []
feat = []        

def tc():
    global xtrain,ytrain,clf
    clf.fit(xtrain,ytrain)

def pl():
    global xtest,ytest,clf
#     print(xtest,ytest)
    yp = clf.predict(xtest)
    return sum(ytest == yp) / float(len(yp)) 

def pred(y):
    global clf
    print(type(y))
    yp = clf.predict(y)
    print(yp)
    global preval
    preval = yp
    prelbl.config(text="PREDICTION IS: "+str(preval[0]))
    
def tp():
    global clf
    print(clf)
    tc()
    acc = pl()
    global accu
    accu = acc
    
def tp2(data):
    global clf
    pred(data)

import tkinter 
tk=tkinter.Tk()
tk.title("CLASSIFICATION")
tk.geometry('900x500+80+80')
tk.configure(background='lightskyblue')
from tkinter import *

var = StringVar()

csvip=Label(tk,text="ENTER CSV PATH",bg='lightskyblue')
csvip.place(x=50,y=50)

csvloc=StringVar()
featlist = StringVar()
predlist = StringVar()
target = StringVar()

def load():
    csvf = csvloc.get()
    global data
    data = pd.read_csv(csvf)
#     print(data)
    ld()

def ld():
    featip=Label(tk,text="ENTER FEATURE LIST",bg='lightskyblue')
    featip.place(x=25,y=100)
    textbox2=Entry(tk,textvariable=featlist,width=75)
    textbox2.place(x=150,y=100)
    tarip=Label(tk,text="ENTER TARGET COLUMN",bg='lightskyblue')
    tarip.place(x=575,y=100)
    textbox3=Entry(tk,textvariable=target)
    textbox3.place(x=725,y=100)
    SubmitButton2=Button(tk,text="ENTER",fg="Red")
    SubmitButton2.place(x=550,y=150)
    SubmitButton2.config(command=enter)
    global var
    var.set("Select a model for prediction")
    ch = ['Logistic Regression','Support Vector Machine','XGBoost','Decision Tree']
    dd = OptionMenu(tk,var,*ch)
    dd.place(x = 300,y = 150)
    
def solve():
    pr = predlist.get()
    pred = pr.split(",")
    predi = [float(i) if i.replace('.','').isdigit() else i for i in pred]
    predict = {feat[i]:predi[i] for i in range(len(feat))}
    pre = pd.DataFrame(predict,index = [0])
    scaler2 = MinMaxScaler()
    global scaler
    pre1 = scaler2.fit_transform(pre)
    pre = scaler.transform(pre)
    tp2(pre)
    
def accuracy():
    global xtrain,xtext,ytrain,ytest
#     print(xtrain,type(xtrain))
    tp()
    global accu
    accu1 = "{:.5f}".format(accu)
    acclbl.config(text="ACCURACY = "+str(accu))
    accu2 = "{:.2f}".format(accu*100)
    acclbl2.config(text="ACCURACY % = "+str(accu2)+"%")
    
def predict():
    predip.config(text="ENTER UNKNOWN FEATURE LIST")
    textbox4=Entry(tk,textvariable=predlist,width=50)
    textbox4.place(x=550,y=250)
    SolveBtn=Button(tk,text="SOLVE",fg="Red")
    SolveBtn.place(x=300,y=300)
    SolveBtn.config(command=solve)
   
    
def enter():
    acclbl.config(text="")
    acclbl2.config(text="")
    global data
    print(data)
    ft = featlist.get()
    global feat
    feat = ft.split(',')
    x = data[feat]
    tar = target.get()
    y = data[tar]
    global xtrain,xtest,ytrain,ytest
    xtrain, xtest, ytrain, ytest = train_test_split(x, y, random_state=0)
#     print(xtest,type(xtest))
    global scaler
    xtrain = scaler.fit_transform(xtrain)
    xtest = scaler.transform(xtest)
#     print(xtest,type(xtest))
#     print(xtrain,type(xtrain))
#     print(var.get())
    v = var.get()
    global clf
    if v == "Logistic Regression":
        clf = clfa
    if v == "Support Vector Machine":
        clf = clfb
    if v == "XGBoost":
        clf = clfc
    if v == "Decision Tree":
        clf = clfd
    AccuracyBtn=Button(tk,text="ACCURACY",fg="Red")
    AccuracyBtn.place(x=150,y=200)
    AccuracyBtn.config(command=accuracy)
    PredictBtn=Button(tk,text="--PREDICTION--",fg="Red")
    PredictBtn.place(x=300,y=200)
#     PredictBtn.config(command=predict)
    predip.config(text="ENTER UNKNOWN FEATURE LIST")
    textbox4=Entry(tk,textvariable=predlist,width=50)
    textbox4.place(x=550,y=250)
    SolveBtn=Button(tk,text="SOLVE",fg="Red")
    SolveBtn.place(x=300,y=300)
    SolveBtn.config(command=solve)
    
textbox=Entry(tk,textvariable=csvloc,width=100)
textbox.place(x=150,y=50)

SubmitButton=Button(tk,text="LOAD",fg="Red")
SubmitButton.place(x=800,y=50)
SubmitButton.config(command=load)

acclbl=Label(tk,text="",bg='lightskyblue')
acclbl.place(x=50,y=250)
acclbl2=Label(tk,text="",bg='lightskyblue')
acclbl2.place(x=50,y=300)

prelbl=Label(tk,text="",bg='lightskyblue',font=16)
prelbl.place(x=300,y=350)
predip=Label(tk,text="",bg='lightskyblue')
predip.place(x=300,y=250)


tk.mainloop()
```
